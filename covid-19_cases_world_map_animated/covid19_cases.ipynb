```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['Province/State', 'Lat', 'Long'])
```

```python
df.head()
```

```python
df = df.melt(id_vars=['Country/Region'], 
        var_name='Date', 
        value_name='Cases')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df['Cases'] != 0]
```

```python
df.head()
```

```python
df.Date.unique()
```

```python
df['Total cases'] = df.groupby(['Country/Region', 'Date'])['Cases'].transform('sum')
```

```python
df_new = df.drop_duplicates(subset=['Country/Region', 'Date'])
```

```python
df_new.head()
```

```python
fig = px.scatter_geo(df_new, locations='Country/Region', locationmode='country names', hover_name='Country/Region',
color_discrete_sequence=px.colors.qualitative.Light24, projection='natural earth',
animation_frame='Date', size='Total cases', size_max=60, width=1000, height=720, title='COVID-19 cases from Jan 22nd to Apr 6th')
fig.show()
import plotly.io as pio
pio.write_html(fig, file='COVID-19_cases.html', auto_open=True)
```

```python
df2 = pd.read_csv('time_series_covid19_confirmed_US.csv')
```

```python
df2.head()
```

```python
for x in df2.columns:
    if x != 'Country_Region' and x != 'Province_State' and x != '4/5/20' and x != 'Lat' and x != 'Long_' and x != 'FIPS':
        df2.drop(x, axis = 1, inplace=True)
```

```python
df2.head()
```

```python
df2.rename(columns={'4/5/20' : 'Cases'}, inplace=True)
```

```python
df2 = df2[df2['Cases'] != 0]
```

```python
px.set_mapbox_access_token(open(".mapbox_token").read())
fig = px.scatter_mapbox(df2, lat='Lat', lon='Long_', size='Cases',
        size_max=80, zoom=3.5, mapbox_style='light', center = {'lat': 37.0902, 'lon': -95.7129},
        color_discrete_sequence=['red'], width=1280, height=720,
        hover_data=['Province_State', 'Lat', 'Long_', 'Cases']
                       )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
pio.write_html(fig, file='COVID-19_cases_USA.html', auto_open=True)
```
