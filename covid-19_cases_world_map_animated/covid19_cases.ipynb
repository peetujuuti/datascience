```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['Province/State', 'Lat', 'Long'])
```

```python
df.head()
```

```python
df = df.melt(id_vars=['Country/Region'], 
        var_name='Date', 
        value_name='Cases')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df['Cases'] != 0]
```

```python
df.head()
```

```python
df.Date.unique()
```

```python
df['Total cases'] = df.groupby(['Country/Region', 'Date'])['Cases'].transform('sum')
```

```python
df_new = df.drop_duplicates(subset=['Country/Region', 'Date'])
```

```python
df_new.head()
```

```python
fig = px.scatter_geo(df_new, locations='Country/Region', locationmode='country names', hover_name='Country/Region',
color_discrete_sequence=px.colors.qualitative.Light24, projection='natural earth',
animation_frame='Date', size='Total cases', size_max=60, width=1000, height=720, title='COVID-19 cases from Jan 22 to Mar 26')
fig.show()
import plotly.io as pio
pio.write_html(fig, file='COVID-19_cases.html', auto_open=True)
```
